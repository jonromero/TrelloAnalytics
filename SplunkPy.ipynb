```python
import pandas as pd
```

```python
filename = '/home/pi/Sources/TrelloAnalytics/data.csv'
```

```python
df = pd.read_csv(open(filename, 'rb'), names=['Date', 'List', 'ID'], parse_dates=['Date'])
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
df['Date'] = df['Date'].dt.date
```

```python
df[df.List.str.strip() != 'Done'].groupby(['Date', 'List'])['List'].count().unstack().plot(kind="bar", title="Tasks")
```

```python
df[df.List.str.strip() == 'Done'].groupby(['Date'])['List'].count().diff().fillna(value=0).mean()
```

```python
df[df.List.str.strip() == 'This Week'].groupby(['Date'])['List'].count().diff().fillna(value=0).plot(kind="bar", title="Tasks Completed")
```

```python
df[df.List.str.strip() != 'Done'].groupby(['Date', 'List'])['List'].count().unstack().plot(kind="bar", title="Tasks Completed")
```
